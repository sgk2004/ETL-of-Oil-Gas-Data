```python
#Dependencies

import pandas as pd
import csv
import os

from datetime import datetime as dt
import numpy as np

from sqlalchemy import create_engine
```

## Data Extraction file 1

## Importing US Crude oil production 2008-2018 in csv format

```python
# Importing US Crude oil production 2008-2018 in csv format
file1_path= os.path.join('Resources','U.S._crude_oil_production.csv')
oil_production_df = pd.read_csv(file1_path)
oil_production_df
```

## Transform the data extracted

```python
# Summarizing the date column month wise to yearly(taking mean for per year)
oil_production_df['year'] = pd.DatetimeIndex(oil_production_df['Month']).year
del oil_production_df['Month']
```

```python
mean_oil_production_df = pd.DataFrame(round(oil_production_df.groupby(['year']).mean(),2))
mean_oil_production_df=mean_oil_production_df.reset_index()
```

```python
oilproducingstates_df=mean_oil_production_df[['year','Alaska','California','Colorado','Kansas','Louisiana',
                                                          'Montana','New Mexico','North Dakota','Oklahoma',
                                                          'Wyoming','Utah','Texas']].copy()
oilproducingstates_df
```

```python
# Transpose the DataFrame.
oilproducingstates_df =oilproducingstates_df.T
oilproducingstates_df
```

```python
#Selecting rows and columns which are needed(for year 2010 to 2014)
oilproducingstates_df=oilproducingstates_df.iloc[1:,2:7]
oilproducingstates_df
```

```python
oilproducingstates_df=oilproducingstates_df.rename(columns={ 2:2010, 3:2011, 4:2012, 5:2013,6:2014})
```

```python
oilproducingstates_df=oilproducingstates_df.reset_index()
```

```python
oilproducingstates_df=oilproducingstates_df.rename(columns={'index':'state'})
```

##  Crude Oil Production of Key States in USA  for the year 2010-2014 

```python
oilproducingstates_df
```

## Normalization 

```python
#df.melt(id_vars=['A'], value_vars=['B', 'C'])
oilproducingstates_df=oilproducingstates_df.melt(id_vars=["state"], value_vars=[2010,2011,2012,2013,2014],
                        var_name="year",value_name="oil")
oilproducingstates_df.head()
```

```python
oilproducingstates_df.info()
```

```python
oilproducingstates_df['year'] =  pd.to_datetime(oilproducingstates_df['year'], format='%Y')
oilproducingstates_df.info()
```

```python
oilproducingstates_df['year']=oilproducingstates_df['year'].dt.year
```

```python
#Final dataframe to load to DB
oilproducingstates_df.head()
```

```python
oilproducingstates_df.info()
```

## Data Extraction file 3

##  GDP Data USA 2010-2014

```python
file3_path= os.path.join('Resources','Energy Census and Economic Data US 2010-2014.csv')
Energy_df = pd.read_csv(file3_path)
Energy_df
```

```python
col_names = Energy_df.columns.tolist()
print(col_names)
```

## Transform the data extracted

```python

states_df=Energy_df.iloc[:,1:2]
states_df
```

```python
GDP_df=pd.DataFrame(Energy_df.filter(regex='^GDP',axis=1))

```

```python
drop_columns_list=Energy_df.filter(regex='Q1|Q2|Q3|Q4').columns
drop_columns_list
```

```python
GDP_df=GDP_df.drop(drop_columns_list, axis=1)
```

```python
GDP_df.head()
```

```python
GDP_df=pd.concat([states_df, GDP_df],axis=1)
```

```python
GDP_df=GDP_df.rename(columns={'State':'state','GDP2010':2010,'GDP2011':2011,'GDP2010':2012,'GDP2012':2010,'GDP2013':2013,'GDP2014':2014})
```

```python
GDP_df.head()
```

```python
GDP_df=GDP_df.melt(id_vars=["state"], value_vars=[2010,2011,2012,2013,2014],
                        var_name="year",value_name="GDP")
GDP_df.head()
```

```python
GDP_df['year'] =  pd.to_datetime(GDP_df['year'], format='%Y')
GDP_df['year']=GDP_df['year'].dt.year
```

```python
GDP_df
```

## Energy Consumption: Fossil Fuel, Natural Gas and LPG 

## Transformation and normalization of DataFrame

```python
Fossilfuel_df=pd.DataFrame(Energy_df.filter(regex='^FossFuel',axis=1))
Naturalgas_df=pd.DataFrame(Energy_df.filter(regex='^NatGasC',axis=1))
LPG_df=pd.DataFrame(Energy_df.filter(regex='^LPGC',axis=1))
```

Fossil Fuel Consumption

```python
Fossilfuel_df=pd.concat([states_df, Fossilfuel_df],axis=1)
Fossilfuel_df
```

```python
Fossilfuel_df=Fossilfuel_df.rename(columns={'State':'state','FossFuelC2010':2010,'FossFuelC2011':2011,'FossFuelC2012':2012,'FossFuelC2013':2013,'FossFuelC2014':2014})
Fossilfuel_df.head()
```

```python
Fossilfuel_df=Fossilfuel_df.melt(id_vars=["state"], value_vars=[2010,2011,2012,2013,2014],
                        var_name="year",value_name="fossil_fuel")
```

```python
Fossilfuel_df['year'] =  pd.to_datetime(Fossilfuel_df['year'], format='%Y')
Fossilfuel_df['year']=Fossilfuel_df['year'].dt.year
Fossilfuel_df.head()
```

```python
Fossilfuel_df
```

Natural Gas Consumption

```python
Naturalgas_df=pd.concat([states_df, Naturalgas_df],axis=1)
Naturalgas_df
```

```python
Naturalgas_df=Naturalgas_df.rename(columns={'State':'s1','NatGasC2010':2010,'NatGasC2011':2011,'NatGasC2012':2012,'NatGasC2013':2013,'NatGasC2014':2014})
Naturalgas_df.head()
```

```python
Naturalgas_df=Naturalgas_df.melt(id_vars=["s1"], value_vars=[2010,2011,2012,2013,2014],
                        var_name="year",value_name="natural_gas")
```

```python
Naturalgas_df['year'] =  pd.to_datetime(Naturalgas_df['year'], format='%Y')
Naturalgas_df['year']=Naturalgas_df['year'].dt.year
```

```python
Naturalgas_df
```

```python
Naturalgas_df.drop(['s1','year'],axis=1,inplace=True)
```

```python
Naturalgas_df
```

LPG Consumption

```python
LPG_df=pd.concat([states_df, LPG_df],axis=1)
LPG_df
```

```python
LPG_df=LPG_df.rename(columns={'State':'state','LPGC2010':2010,'LPGC2011':2011,'LPGC2012':2012,'LPGC2013':2013,'LPGC2014':2014})
LPG_df.head()
```
